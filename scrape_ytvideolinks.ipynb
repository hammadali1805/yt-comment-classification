```python
import yt_dlp

def get_channel_videos(channel_url):
    # Create an instance of the yt-dlp object
    ydl = yt_dlp.YoutubeDL()
    
    # Extract information from the channel URL
    info_dict = ydl.extract_info(channel_url, download=False)
    
    # Check if the URL is a playlist
    if 'entries' in info_dict:
        videos = info_dict['entries']
    else:
        videos = [info_dict]
    
    # Extract video details
    video_details = []
    for video in videos:
        video_details.append({
            'title': video.get('title'),
            'id': video.get('id'),
            'url': video.get('url'),
            'upload_date': video.get('upload_date')
        })
    
    return video_details

# Replace with the desired YouTube channel URL
channel_url = 'https://www.youtube.com/@CarryMinati'
video_details = get_channel_videos(channel_url)

# for video in video_details:
#     print(f"Title: {video['title']}")
#     print(f"ID: {video['id']}")
#     print(f"URL: {video['url']}")
#     print(f"Upload Date: {video['upload_date']}\n")

len(video_details)
```

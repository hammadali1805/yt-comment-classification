```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

df = pd.read_csv('shortDataset.csv')
```

```python
sentences = df['comment'].to_list()
```

```python
# Initialize the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Fit-transform the corpus to compute the TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(sentences)

# Get the feature names (words) from the vectorizer
feature_names = vectorizer.get_feature_names_out()

# Convert the TF-IDF matrix to a dense format (DataFrame) for easy inspection
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

# Print the TF-IDF values for each word in each document
tfidf_df
```

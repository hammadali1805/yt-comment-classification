```python
from transformers import BertTokenizer, BertModel
import torch
import pandas as pd
```

```python
df = pd.read_csv('shortDataset.csv')
words =  (' '.join(df['comment'].to_list())).split(' ')
words = set(words)
words.remove('')
len(words)
```

```python
# Load pre-trained BERT model and tokenizer
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
```

```python
# Create a dictionary to store word embeddings
word_embeddings = {}

# Set the model to evaluation mode
model.eval()

# Extract embeddings for each word in the vocabulary
with torch.no_grad():
    i=0
    for word in words:
        # Tokenize the word and convert to input IDs
        input_ids = tokenizer(word, return_tensors="pt")["input_ids"]

        # Pass the input IDs through BERT
        outputs = model(input_ids)

        # Extract the last hidden state (embeddings) for the word
        # The embedding for the word is the first token's vector (excluding [CLS] and [SEP])
        embedding = outputs.last_hidden_state[0][1].numpy()

        # Store the embedding in the dictionary
        word_embeddings[word] = embedding
        i+=1
        print(f'{i}/26409')
```

```python
import pickle

# Save the word embeddings dictionary as a pickle file
with open("bert_word_embeddings_shortDataset.pkl", "wb") as file:
    pickle.dump(word_embeddings, file)
```
